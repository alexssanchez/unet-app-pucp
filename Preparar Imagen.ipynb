```
import re
import glob
# Imagenes de Entrenamiento
paths_tr = glob.glob("Task04_Hippocampus/imagesTr/*.nii.gz")
numbers_tr = list(map(lambda x: str(int(re.search(r'\d+', x.split('hippocampus_')[1]).group(0))), paths_tr))

# Imagenes de Pueba
paths_ts = glob.glob("Task04_Hippocampus/imagesTs/*.nii.gz")
numbers_ts = list(map(lambda x: str(int(re.search(r'\d+', x.split('hippocampus_')[1]).group(0))), paths_ts))

paths_ls = glob.glob("Task04_Hippocampus/labelsTr/*.nii.gz")
numbers_ts = list(map(lambda x: str(int(re.search(r'\d+', x.split('hippocampus_')[1]).group(0))), paths_ls))

# Comandos linux para crear directorios
!mkdir Hippocampus
!mkdir Hippocampus/imagesTr
!mkdir Hippocampus/imagesTs
!mkdir Hippocampus/labelsTr
```

### Conversion de imagenes

Las imagenes estan en formato nii, se utilizo la libreria [nii2png.py](https://github.com/alexlaurence/NIfTI-Image-Converter) para convertirlas. para mayor info revisar [NIfTI-Image-Converter](https://github.com/alexlaurence/NIfTI-Image-Converter).

```
import nii2png as cvt

#extrae las imagenes de training y las guarda en la carpeta
for i in range(len(paths_tr)):
    cvt.convert(paths_tr[i],'Hippocampus/imagesTr/')
```

```
#extrae las imagenes de test y las guarda en la carpeta
for j in range(len(paths_ts)):
  cvt.convert(paths_ts[j],'Hippocampus/imagesTs/')
```

```
#extrae las imagenes de test y las guarda en la carpeta
for k in range(len(paths_ls)):
  cvt.convert(paths_ls[k],'Hippocampus/labelsTr/')
```
